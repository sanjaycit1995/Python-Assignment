3 Compare and contrast mutable and immutable objects in Python with examples.

Ans-Mutable — the object's contents can be changed after creation, without changing its identity (memory address).

my_list = [1, 2, 3]
print(id(my_list))       # e.g. 140234567890
my_list.append(4)        # modifies in place
print(my_list)           # [1, 2, 3, 4]
print(id(my_list))       # same id as before — same object

Immutable — once created, the object's value can never be changed. Any "modification" actually creates a new object.

my_str = "hello"
print(id(my_str))
my_str = my_str + " world"   # this doesn't modify the original string —
                               # it creates a brand new string object
print(my_str)                 # "hello world"
print(id(my_str))             # different id — it's a new object
