2 Explain the difference between mutable and immutable data types with examples.

Ans- Mutable Data Types

A mutable object allows you to change, add, or remove its elements without creating a new object.

Examples of Mutable Data Types
List (list)
Dictionary (dict)
Set (set)
Bytearray (bytearray)

fruits = ["Apple", "Mango", "Orange"]

fruits[1] = "Banana"

print(fruits)

An immutable object cannot be changed after it is created. If you try to modify it, Python creates a new object or raises an error.

Examples of Immutable Data Types
String (str)
Tuple (tuple)
Integer (int)
Float (float)
Boolean (bool)
Frozenset (frozenset)

name = "Sanjay"

name = "Kumar"

print(name)

8 What is a hash table, and how does it relate to dictionaries in Python?


A hash table is a data structure that stores data in the form of key-value pairs. It uses a hash function to convert a key into a unique hash value (an index), allowing data to be stored and retrieved very quickly.

A hash table provides fast lookup, insertion, and deletion, usually in O(1) (constant time) on average.

A Python dictionary (dict) is implemented using a hash table.

The key is passed through a hash function.
The hash function generates a hash value.
The hash value determines where the key-value pair is stored in memory.
When you access a value using its key, Python computes the hash again and quickly retrieves the value.

This is why dictionary operations such as lookup, insertion, and deletion are very fast.
